```python
#Creating a function that will setup the twitter API and allow me to feed a screen name to extract tweets out of
def twitter_setup():

    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_key,access_secret)

    api = tweepy.API(auth)
    return api
```

```python
tweet_extract = twitter_setup()

#Setting up timeline of user, TomiLahren, who is a divisive political commentary, which I will use later on
#in the project to analyze the polarity of her tweets
tweets = tweet_extract.user_timeline(screen_name="TomiLahren", count=200)

#loop to go through the tweets that were extracted
for tweet in tweets[:200]:
    print(tweet.text)
```

```python
#Creating a dataframe so we can later use to analyze the tweets
data = pd.DataFrame(data=[tweet.text for tweet in tweets], columns=['Tweets'])

# Displaying the pandas dataframe that now has the contents of the first 20 tweets
display(data.head(20))
```

```python
#Now we are adding Length, Date, number of likes, and number of retweeets, into the dataframe
data['len']  = np.array([len(tweet.text) for tweet in tweets])
data['Date'] = np.array([tweet.created_at for tweet in tweets])
data['Likes']  = np.array([tweet.favorite_count for tweet in tweets])
data['RTs']    = np.array([tweet.retweet_count for tweet in tweets])
```

```python
# Display of first 10 elements from dataframe:
display(data.head(20))
```

```python
#Function used to cleanup the tweet to remove URL and special characters
def cleanup_tweet(tweet):
  
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())

```

```python
#Sentiment Analysis function, that is using the TweetBlob library, that uses NLP, to classify the polarity of the tweets
def analyze_sentiment(tweet):
    
    analysis = TextBlob(cleanup_tweet(tweet))
    if analysis.sentiment.polarity > 0:
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
```

```python
# Adding a new column to the data frame, that will include the polarity value of the tweet
#1 = positive, 0 = neutral, and -1 = negative tweet
data['Sentiment'] = np.array([ analyze_sentiment(tweet) for tweet in data['Tweets'] ])

display(data.head(20))
```

```python
#Extracting the tweets that had the most likes and most retweets

mostlikes_max = np.max(data['Likes'])
retweet_max  = np.max(data['RTs'])

mostlikes = data[data.Likes == mostlikes_max].index[0]
retweet  = data[data.RTs == retweet_max].index[0]

# Max FAVs:
print("The tweet with more likes is: \n{}".format(data['Tweets'][mostlikes]))
print("Number of likes: {}".format(mostlikes_max))
print("The most popular tweet's polarity is: \n{}""".format(analyze_sentiment(format(data['Tweets'][mostlikes]))))

# Max RTs:
print("\nThe tweet with more retweets is: \n{}".format(data['Tweets'][retweet]))
print("Number of retweets: {}".format(retweet_max))
print("The most reweeted tweet's polarity is: \n{}""".format(analyze_sentiment(format(data['Tweets'][retweet]))))
```

```python
#We can see that the most popular tweets were negative
#the most popular retweets were neutral
```

```python
#Now we are separating the positive, neutral, and negative tweets, so we can figure out the percentage
#of polarity of the 200 tweets

positive_tweets = [ tweet for index, tweet in enumerate(data['Tweets']) if data['Sentiment'][index] > 0]
neutral_tweets = [ tweet for index, tweet in enumerate(data['Tweets']) if data['Sentiment'][index] == 0]
negative_tweets = [ tweet for index, tweet in enumerate(data['Tweets']) if data['Sentiment'][index] < 0]
```

```python
#Here we are printing the percentages of the tweets

print("Percentage of positive tweets: {}%".format(len(positive_tweets)*100/len(data['Tweets'])))
print("Percentage of neutral tweets: {}%".format(len(neutral_tweets)*100/len(data['Tweets'])))
print("Percentage de negative tweets: {}%".format(len(negative_tweets)*100/len(data['Tweets'])))
```
